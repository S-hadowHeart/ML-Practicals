# Tutorial 0: Describe Any 10 AI Libraries

This tutorial provides an overview of 10 popular AI libraries, focusing on their functionalities, applications, and key features.

---

## 1. TensorFlow
**Description:** TensorFlow is an open-source machine learning framework developed by Google. It is widely used for building and deploying machine learning models, especially deep learning neural networks.

**Use Cases:** Image recognition, natural language processing, speech recognition, and reinforcement learning.

**Key Features:** 
- Supports both CPU and GPU computing.
- Flexible architecture for deployment on various platforms (mobile, edge, cloud).
- Extensive ecosystem including TensorBoard for visualization and TensorFlow Lite for mobile.

---

## 2. PyTorch
**Description:** PyTorch is an open-source deep learning framework developed by Facebook's AI Research lab. It is known for its dynamic computation graph and ease of use.

**Use Cases:** Research prototyping, computer vision, natural language processing, and reinforcement learning.

**Key Features:** 
- Dynamic computation graph for flexible model building.
- Strong GPU acceleration.
- Rich ecosystem including TorchVision and TorchText.

---

## 3. Keras
**Description:** Keras is a high-level neural networks API, written in Python and capable of running on top of TensorFlow, Theano, or CNTK. It is designed for fast experimentation.

**Use Cases:** Rapid prototyping of deep learning models, image and text classification.

**Key Features:** 
- User-friendly and modular.
- Supports convolutional and recurrent networks.
- Easy integration with TensorFlow backend.

---

## 4. OpenCV
**Description:** OpenCV (Open Source Computer Vision Library) is an open-source computer vision and machine learning software library.

**Use Cases:** Real-time image processing, object detection, facial recognition, and augmented reality.

**Key Features:** 
- Extensive algorithms for image and video analysis.
- Supports multiple programming languages (C++, Python, Java).
- Optimized for real-time applications.

---

## 5. Scikit-Learn
**Description:** Scikit-Learn is a simple and efficient tool for data mining and data analysis built on top of Python libraries like NumPy and SciPy.

**Use Cases:** Classification, regression, clustering, dimensionality reduction, and model selection.

**Key Features:** 
- Easy-to-use API.
- Wide range of machine learning algorithms.
- Integration with other Python scientific libraries.

---

## 6. NLTK (Natural Language Toolkit)
**Description:** NLTK is a leading platform for building Python programs to work with human language data.

**Use Cases:** Text processing, tokenization, parsing, semantic reasoning, and sentiment analysis.

**Key Features:** 
- Comprehensive libraries for linguistic data processing.
- Support for classification, tokenization, stemming, tagging.
- Large collection of corpora and lexical resources.

---

## 7. SpaCy
**Description:** SpaCy is an open-source software library for advanced natural language processing in Python.

**Use Cases:** Named entity recognition, part-of-speech tagging, dependency parsing, and text classification.

**Key Features:** 
- Industrial-strength NLP with fast performance.
- Pre-trained models for multiple languages.
- Easy integration with deep learning frameworks.

---

## 8. GPT-3 (OpenAI)
**Description:** GPT-3 is a state-of-the-art language generation model developed by OpenAI, capable of understanding and generating human-like text.

**Use Cases:** Text generation, translation, summarization, chatbots, and code generation.

**Key Features:** 
- Large-scale transformer architecture.
- Few-shot and zero-shot learning capabilities.
- API access for easy integration.

---

## 9. IBM Watson
**Description:** IBM Watson is a suite of AI services, applications, and tools designed for enterprises to build and deploy AI-powered solutions.

**Use Cases:** Natural language understanding, speech-to-text, visual recognition, and AI-powered chatbots.

**Key Features:** 
- Cloud-based AI services.
- Pre-built models and customizable AI tools.
- Integration with enterprise systems.

---

## 10. RapidMiner
**Description:** RapidMiner is a data science platform that provides an integrated environment for machine learning, data preparation, and model deployment.

**Use Cases:** Predictive analytics, customer segmentation, fraud detection, and text mining.

**Key Features:** 
- Visual workflow designer.
- Supports automated machine learning.
- Integration with various data sources and formats.

2. Pie chart (cagegorial data)
    A pie chart represents the parts of a whole entity.

```python
df['class'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
df['source_city'].value_counts().plot(kind='pie', autopct='%.2f')
```

3. Donut plots (used for categorial data)
    Donut plots are used to show the proportion of each category in relation to the whole.

```python
plt.pie(df['source_city'].value_counts(), labels=df['source_city'].unique(), autopct='%.2f', wedgeprops=dict(width=.5))
```

4. Histogram (used for numerical data) (Contiguous variables)
  Histograms are used when you have a contiguous variable, and you want to visualize its distribution.
