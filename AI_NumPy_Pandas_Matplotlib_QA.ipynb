# Exercise 1: Describe Any 10 AI Libraries

**Description:** Overview of popular AI libraries with use cases and key features.


### 1. TensorFlow
**Description:** An open-source deep learning framework developed by Google.  
**Use Cases:** Deep learning, neural networks, computer vision, NLP.  
**Key Features:** Scalable, GPU/TPU support, production-ready models.

### 2. PyTorch
**Description:** An open-source deep learning framework by Meta (Facebook).  
**Use Cases:** Research, deep learning, dynamic neural networks.  
**Key Features:** Dynamic computation graph, easy debugging, Pythonic.

### 3. Keras
**Description:** High-level neural network API running on TensorFlow.  
**Use Cases:** Rapid deep learning prototyping.  
**Key Features:** Simple syntax, modular, user-friendly.

### 4. OpenCV
**Description:** Open-source computer vision library.  
**Use Cases:** Image processing, face detection, video analysis.  
**Key Features:** Real-time processing, multi-language support.

### 5. Scikit-Learn
**Description:** Machine learning library for classical ML algorithms.  
**Use Cases:** Classification, regression, clustering.  
**Key Features:** Simple API, well-documented, efficient.

### 6. NLTK
**Description:** Natural Language Toolkit for text processing.  
**Use Cases:** Tokenization, stemming, text analysis.  
**Key Features:** Educational, rich NLP tools.

### 7. SpaCy
**Description:** Industrial-strength NLP library.  
**Use Cases:** Named entity recognition, text classification.  
**Key Features:** Fast, accurate, production-ready.

### 8. GPT-3 (OpenAI)
**Description:** Large language model for text generation.  
**Use Cases:** Chatbots, content creation, Q&A systems.  
**Key Features:** Few-shot learning, natural language understanding.

### 9. IBM Watson
**Description:** AI platform by IBM for enterprise AI.  
**Use Cases:** Chatbots, analytics, healthcare AI.  
**Key Features:** Cloud-based, business integration.

### 10. RapidMiner
**Description:** Data science platform for analytics.  
**Use Cases:** Data mining, predictive analytics.  
**Key Features:** GUI-based, no-code/low-code.

# Exercise 3: Explore Pandas Library

```

import pandas as pd
import numpy as np

# Series
s = pd.Series([10,20,30])
s_custom = pd.Series([100,200,300], index=['a','b','c'])
s_custom.values, s_custom.index

# DataFrame from dict
df = pd.DataFrame([{'Name':'A','Age':20},{'Name':'B','Age':21}])

# Iris dataset
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
iris = pd.read_csv(url)

iris.shape
iris.head()
iris.tail()
iris.index

# Rename columns
iris.rename(columns={
    'sepal_length':'Sepal_Length',
    'sepal_width':'Sepal_Width',
    'petal_length':'Petal_Length',
    'petal_width':'Petal_Width'
}, inplace=True)

# Replace categories
iris['species'] = iris['species'].replace({
    'Iris-setosa':'setosa',
    'Iris-versicolor':'versicolor',
    'Iris-virginica':'virginica'
})

iris.describe()

# Extract column
iris['Petal_Length']

# Add and drop column
iris['Size'] = np.where(iris['Petal_Length']>5,'Large','Small')
iris.drop('Size', axis=1, inplace=True)

# Row selection
iris.iloc[0]
iris.iloc[50:55]

# Conditional filter
iris[(iris['species']=='virginica') & (iris['Petal_Length']>6)]

# Duplicates, info, nulls
iris.duplicated().sum()
iris.info()
iris.isnull().sum()
```
