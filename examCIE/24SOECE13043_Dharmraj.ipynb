```python
import pandas as pd
import numpy as np
```

```python
house_data = pd.read_csv("../datasets/House Price India.csv")
house_data.head()
```

```python
house_data.isnull().sum()
```

```python
house_data.fillna(house_data["Distance from the airport"].mean(),inplace=True)
house_data.fillna(house_data["Price"].mean(),inplace=True)
```

```python
x = house_data.iloc[:,2:-1]
y = house_data.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,random_state=42 , test_size=0.2)
```

```python
from sklearn.linear_model import Lasso
```

```python
model = Lasso(alpha=0.1)
model.fit(x_train,y_train)
y_predict = model.predict(x_test)
```

```python
model.coef_
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
std_mark = pd.read_csv("../datasets/std_marks_data.csv")
```

```python
std_mark.isnull().sum()
```

```python
std_mark.fillna(std_mark["hours"].mean(),inplace=True)
```

```python
x = std_mark.iloc[:,:-1]
y = std_mark.iloc[:,-1]
```

```python
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_scaler , y , random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train , y_train)
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test , y_predict)
```
