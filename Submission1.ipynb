```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

nltk.download('punkt')

text = "NLTK is very useful for text processing"
words = word_tokenize(text)

print(words)
```

```python
from nltk.tokenize import sent_tokenize

text = "NLTK is easy. It is useful for NLP."
sentences = sent_tokenize(text)

print(sentences)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')

text = "This is a simple example of NLTK"
stop_words = set(stopwords.words('english'))

words = word_tokenize(text)
filtered_words = [w for w in words if w.lower() not in stop_words]

print(filtered_words)
```

### Find diff. b/w python list and np array



* Python list:
  * Python lists are excellent, general-purpose containers. They can be “heterogeneous”, meaning that they can contain elements of a variety of types, and they are quite fast when used to perform individual operations on a handful of elements.

    Depending on the characteristics of the data and the types of operations that need to be performed, other containers may be more appropriate; by exploiting these characteristics, we can improve speed, reduce memory consumption, and offer a high-level syntax for performing a variety of common processing tasks. NumPy shines when there are large quantities of “homogeneous” (same-type) data to be processed on the CPU.

* Numpy array:
  * Most NumPy arrays have some restrictions. For instance:

    All elements of the array must be of the same type of data.

    Once created, the total size of the array can’t change.

    The shape must be “rectangular”, not “jagged”; e.g., each row of a two-dimensional array must have the same number of columns.

    When these conditions are met, NumPy exploits these characteristics to make the array faster, more memory efficient, and more convenient to use than less restrictive data structures.

### Q. Find the diff. b/w list(1D), numpy array (1D), panda series

#### Library

  * Python List (1D): Built-in Python data structure
  * NumPy Array (1D): Provided by NumPy library
  * Pandas Series: Provided by Pandas library

#### Data Type

  * Python List (1D): Can store different data types
  * NumPy Array (1D): Stores elements of the same data type
  * Pandas Series: Can store different data types

#### Indexing

  * Python List (1D): Uses integer index (0, 1, 2, …)
  * NumPy Array (1D): Uses integer index (0, 1, 2, …)
  * Pandas Series: Uses custom index or labels

#### Performance

  * Python List (1D): Slower in numerical operations
  * NumPy Array (1D): Very fast due to optimized C implementation
  * Pandas Series: Fast, built on top of NumPy

#### Mathematical Operations

  * Python List (1D): Limited support for mathematical operations
  * NumPy Array (1D): Supports vectorized mathematical operations
  * Pandas Series: Supports vectorized mathematical operations

#### Missing Values

  * Python List (1D): No built-in support for missing values
  * NumPy Array (1D): Limited support using nan
  * Pandas Series: Strong support using NaN

#### Memory Usage

  * Python List (1D): Uses more memory
  * NumPy Array (1D): Memory efficient
  * Pandas Series: Uses more memory than NumPy arrays

#### Use Case

  * Python List (1D): General-purpose programming
  * NumPy Array (1D): Scientific and numerical computing
  * Pandas Series: Data analysis and data manipulation
