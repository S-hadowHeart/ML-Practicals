1. K- Means Cluistering without ELbow Method

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
X = df[['sepal length (cm)','sepal width (cm)']].values
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
Kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
y_Kmeans = Kmeans.fit_predict(x_scaled)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,1],hue=y_Kmeans,palette='viridis',s=100)
plt.scatter(Kmeans.cluster_centers_[:,0],Kmeans.cluster_centers_[:,1],s=300,c='red',marker='X',label='Centroids')

plt.xlabel('Sepal Length (standradized)')
plt.ylabel('Sepal width (Standardized)')
plt.title('K-Means Clustering of Iris Dataset (k=3)')
plt.legend()
plt.show()
```

2. K-Means clustering with e

```python
# Compute WCSS for diffrent values of K
wcss = []
k_values = range(1,11)

for k in k_values:
    Kmeans = KMeans(n_clusters=k,random_state=42,n_init=10)
    Kmeans.fit(X)
    wcss.append(Kmeans.inertia_)
```
